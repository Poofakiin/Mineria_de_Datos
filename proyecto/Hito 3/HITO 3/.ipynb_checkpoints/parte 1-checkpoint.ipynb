##1. Introducción


La música es una vía de expresión del ser humano y nunca ha dejado de estar presente, pero la forma de su distribución sí ha cambiado. El mercado mundial de los servicios de *streaming* de música ha crecido muchísimo en los últimos años, y representan una proporción cada vez mayor del mercado de música en general. Por lo tanto, da abundante espacio al estudio y análisis de la música, ya que se puede estudiar el objeto como tal, además de la teoría ya existente, más aún cuando los servicios de *streaming* optan por integrar funcionalidades automatizadas en sus aplicaciones.

Además, el ser humano se dedica a clasificar todo lo que existe, incluyendo la música. Por ello, es interesante estudiar la categorización que el mismo ser humano ha creado - el género. Es común pensar que las canciones se definen en términos de género por la forma en que suenan más que de lo que hablan. Por lo tanto, se puede estudiar si esta hipótesis es certera, analizando aspectos técnicos para el sonido (lo que suena) y las letras (lo que dicen) de las canciones para los temas de los que tratan.

La utilidad del estudio de la clasificación de las canciones recae en los algoritmos de recomendación, pues el programador de estos debe decidir qué criterios utilizar para ver si dos canciones son similares. A su vez, estos pueden ser útiles para distintos propósitos:

- Generación automática de listas de reproducción.
- Complementar el desarrollo de aplicaciones que detectan canciones a partir de un segmento de audio.
- Encontrar canciones similares a otras para integrar la banda sonora en una película o serie.

##3. Preguntas y problemas

Un sistema de recomendación es un algoritmo, y debe entonces el programador considerar ciertos parámetros para tomar decisiones sobre qué tanto se parecen dos canciones. Luego, es natural preguntarse cuáles son esos parámetros. Una opción puede ser considerar los atributos técnicos (lo que suena), como bailabilidad, acústica, instrumentalidad, entre otros. Otra opción puede ser considerar la letra de una canción (lo que dice). Se hipotetiza que la primera opción es la que arroja resultados más certeros para identificar similitudes entre canciones. En este proyecto se busca estudiar si efectivamente es más lógico comparar canciones por cómo suenan que por lo que dicen, para efectos de la programación de un algoritmo de recomendación.

Con esto nacen las preguntas:
1. ¿Los aspectos técnicos de una canción son determinantes en su género?
2. ¿La letra de una canción es determinante en su género?

Para ello, es de interés estudiar la clasificación natural que el ser humano genera para las canciones, que sería el género. De esta forma, sería útil estudiar el atributo género, a modo de saber si es más fácil clasificar una canción por sus atributos técnicos o por su letra.

Por lo anterior, se busca responder la pregunta con técnicas de Minería de Datos.
*   Problema de clasificación 1: Generar un modelo (I) que pueda predecir el género de una canción basándose en los aspectos técnicos como bailabilidad, volumen, energía, entre otros.
*   Problema de clasificación 2: Generar un modelo (II) que pueda predecir el género de una canción basándose en las palabras contenidas en la letra de ella.

Asimismo, sería interesante conocer si lo que llamamos género efectivamente tiene una relación matemática con los aspectos técnicos de las canciones. Es decir, si efectivamente lo que percibimos como pop, jazz, rock tienen aspectos técnicos que los distinguen unos de otros.

Con ello nace la pregunta:

3. ¿Las canciones se agrupan de forma natural en los géneros que conocemos, considerando sus aspectos técnicos, como bailabilidad, volumen, energía, entre otros?

* Problema de clustering: Generar un modelo (III) que pueda agrupar las canciones a partir de los aspectos técnicos como bailabilidad, volumen, energía, entre otros.

Dados los problemas y preguntas planteadas, se tiene que las preguntas principales que guían todo el proyecto son:
¿El género de una canción está más ligado al sonido (lo que suena) o a la verbalización (lo que dice) de una canción?
¿La programación de un algoritmo de recomendación debiese considerar la letra (lo que dice) o aspectos técnicos (lo que suena) para considerar canciones como similares?

Luego, se puede comparar el éxito de los modelos de clasificación y aproximarse a responder cuál de los dos es más determinante para clasificar un género, que se justifica si el modelo de clustering arroja resultados positivos.

Si el modelo (I) tiene un mejor desempeño que el modelo (II), se podría confirmar la hipótesis de que los atributos técnicos son más decisivos a la hora de clasificar una canción. Por lo tanto, se concluiría que estos son los atributos que un programador de un sistema de recomendación debiese tomar en consideración con mayor importancia.

##5. Resultados preliminares

Para este Hito, se ha desarrollado la primera propuesta, esto es, clasificación para predecir género a partir de aspectos técnicos.


### Primera propuesta

Se crea el dataframe. Los datos afortunadamente ya vienen en su forma normalizada.

```python
df_tecnico = df[['genre', 'danceability', 'loudness', 'acousticness', 'instrumentalness', 'valence', 'energy']]
df_tecnico.head()
```

Se separan los datos en conjuntos de evaluación y entrenamiento, en una proporcion 25/75.

```python
from sklearn.model_selection import train_test_split

y = df_tecnico['genre'].values
X_train, X_test, y_train, y_test = train_test_split(df_tecnico, y, test_size=0.25, stratify=y)
```

Se observa nuevamente el desbalance de los datos bajo los que se quiere entrenar.

```python
import numpy as np
np.unique(y_train,return_counts=True)
```

Se aplican técnicas de subsampling y de oversampling al conjunto de entrenamiento, para poder intentar generar modelos con ambas y ver cuál es la mejor.

```python
#oversampling sobre el grupo de training
X_train = X_train.reset_index(drop=True)
idx = np.random.choice(X_train[X_train['genre'] == "country"].index, size=1197)
data_oversampled = pd.concat([X_train, X_train.iloc[idx]])

idx = np.random.choice(X_train[X_train['genre'] == "blues"].index, size=1828)
data_oversampled = pd.concat([data_oversampled, X_train.iloc[idx]])

idx = np.random.choice(X_train[X_train['genre'] == "rock"].index, size=2256)
data_oversampled = pd.concat([data_oversampled, X_train.iloc[idx]])

idx = np.random.choice(X_train[X_train['genre'] == "jazz"].index, size=2397)
data_oversampled = pd.concat([data_oversampled, X_train.iloc[idx]])

idx = np.random.choice(X_train[X_train['genre'] == "reggae"].index, size=3407)
data_oversampled = pd.concat([data_oversampled, X_train.iloc[idx]])

idx = np.random.choice(X_train[X_train['genre'] == "hip hop"].index, size=4603)
data_oversampled = pd.concat([data_oversampled, X_train.iloc[idx]])

print("Data oversampled on genre`s classes")
print(data_oversampled['genre'].value_counts())
print()


#subsampling sobre el grupo de training
idx = np.random.choice(X_train.loc[X_train.genre == "pop"].index, size=4603, replace=False)
data_subsampled = X_train.drop(X_train.iloc[idx].index)

idx = np.random.choice(X_train.loc[X_train.genre == "country"].index, size=3406, replace=False)
data_subsampled = data_subsampled.drop(X_train.iloc[idx].index)

idx = np.random.choice(X_train.loc[X_train.genre == "blues"].index, size=2775, replace=False)
data_subsampled = data_subsampled.drop(X_train.iloc[idx].index)

idx = np.random.choice(X_train.loc[X_train.genre == "rock"].index, size=2347, replace=False)
data_subsampled = data_subsampled.drop(X_train.iloc[idx].index)

idx = np.random.choice(X_train.loc[X_train.genre == "jazz"].index, size=2206, replace=False)
data_subsampled = data_subsampled.drop(X_train.iloc[idx].index)

idx = np.random.choice(X_train.loc[X_train.genre == "reggae"].index, size=1196, replace=False)
data_subsampled = data_subsampled.drop(X_train.iloc[idx].index)
print("Data subsampled on genre`s classes")
print(data_subsampled['genre'].value_counts())
```

Se generan los conjuntos de entrenamiento finales.

```python
# datos para testing
X_test = X_test[X_test.columns.difference(["genre"])].values
y_test = y_test

# datos entrenamiento "originales"
X_train = X_train[X_train.columns.difference(["genre"])].values 
y_train = y_train  

df_tecnico[df_tecnico.columns.difference(['genre'])].values
# datos entrenamiento "oversampleados" 
X_over = data_oversampled[data_oversampled.columns.difference(["genre"])].values
y_over = data_oversampled["genre"].values

# datos entrenamiento "subsampleados"
X_subs = data_subsampled[data_subsampled.columns.difference(["genre"])].values
y_subs = data_subsampled["genre"].values
```

Ahora se aplicaran 3 modelos de clasificacion a los datos seleccionados, SVC, K Nearest Neighbors y Decision Tree. Para estos dos últimos dos se ocupa GridSearchCV con el fin de encontrar los hiperparámetros óptimos para estos modelos.


###Modelo SVC

Se genera el modelo con data original, under y oversampling.

```python
from sklearn.svm import SVC
#Comparacion de modelos de SVC

#Original data
clf_SVC_o=SVC()
clf_SVC_o.fit(X_train, y_train)
y_pred_SVC_o = clf_SVC_o.predict(X_test)

#Subsampled data
clf_SVC_sub=SVC()
clf_SVC_sub.fit(X_subs, y_subs)
y_pred_SVC_sub = clf_SVC_sub.predict(X_test)

#Oversampled data
clf_SVC_over=SVC()
clf_SVC_over.fit(X_over, y_over)
y_pred_SVC_over = clf_SVC_over.predict(X_test)
```

Se obtienen los siguientes resultados:

```python
#Resultados
from sklearn.metrics import classification_report
print("SVC models: ")
print("Original data:")
print(classification_report(y_test, y_pred_SVC_o))
print("Subsample Data:")
print(classification_report(y_test, y_pred_SVC_sub))
print("Oversample Data:")
print(classification_report(y_test, y_pred_SVC_over))
```

Valores agregados por weighted average, en orden, precision, recall y f1-score:

*   Original: 0.41, 0.38, 0.36
*   Subsampling: 0.33, 0.33, 0.29
*   Oversampling: 0.37, 0.35, 0.32






### K Nearest Neighbors

Se genera el modelo con data original, under y oversampling.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier

#KNN parameters
tuned_parameters = {'n_neighbors': [ 2, 3, 4, 5, 10], 
                    'weights': ['uniform','distance']}
score = 'precision' 

#KNN Original values
clf_kmeans_o = GridSearchCV(KNeighborsClassifier(), 
                   param_grid=tuned_parameters, 
                   cv=5,
                   scoring=score)
clf_kmeans_o.fit(X_train, y_train)
y_pred_kmeans_o = clf_kmeans_o.predict(X_test)


#KNN Subsampled values
clf_kmeans_s = GridSearchCV(KNeighborsClassifier(), 
                   param_grid=tuned_parameters, 
                   cv=5,
                   scoring=score)
clf_kmeans_s.fit(X_subs, y_subs)
y_pred_kmeans_s = clf_kmeans_s.predict(X_test)

#KNN Oversampled values
clf_kmeans_over = GridSearchCV(KNeighborsClassifier(), 
                   param_grid=tuned_parameters, 
                   cv=5,
                   scoring=score)
clf_kmeans_over.fit(X_over, y_over)
y_pred_kmeans_over = clf_kmeans_over.predict(X_test)
```

Se obtienen los siguientes resultados:

```python
#Resultados
print("Mejor combinación de parámetros:")
print(clf_kmeans_over.best_params_)
print("KMeans models: ")
print("Original data:")
print(classification_report(y_test, y_pred_kmeans_o))
print("Subsample Data:")
print(classification_report(y_test, y_pred_kmeans_s))
print("Oversample Data:")
print(classification_report(y_test, y_pred_kmeans_over))
```

Valores agregados por weighted average, en orden, precision, recall y f1-score:

*   Original: 0.30, 0.28, 0.27
*   Subsampling: 0.28, 0.24, 0.23
*   Oversampling: 0.28, 0.28, 0.27

####Modelo Decision Tree

Se genera el modelo con data original, under y oversampling.

```python
from sklearn.tree import DecisionTreeClassifier
#Comparacion de modelos de Decision Tree

#Decision Tree parameters
tuned_parameters = {'criterion': ['gini','entropy'], 
                    'max_depth': [3,5,7,10]}
score = 'f1' 

#Decision Tree Original values
clf_decision_o = GridSearchCV(DecisionTreeClassifier(), 
                   param_grid=tuned_parameters, 
                   cv=10,
                   scoring=score)

clf_decision_o.fit(X_train, y_train)
y_pred_decision_o = clf_decision_o.predict(X_test)

#Decision Tree Subsampled values
clf_decision_sub = GridSearchCV(DecisionTreeClassifier(), 
                   param_grid=tuned_parameters, 
                   cv=10,
                   scoring=score)

clf_decision_sub.fit(X_subs, y_subs)
y_pred_decision_sub = clf_decision_sub.predict(X_test)

#Decision Tree Oversampled values
clf_decision_over = GridSearchCV(DecisionTreeClassifier(), 
                   param_grid=tuned_parameters, 
                   cv=10,
                   scoring=score)

clf_decision_over.fit(X_over, y_over)
y_pred_decision_over = clf_decision_over.predict(X_test)
```

Se obtienen los siguientes resultados:

```python
#Resultados
print("Mejor combinación de parámetros:")
print(clf_decision_o.best_params_)
print("DecisionTree models: ")
print("Original data:")
print(classification_report(y_test, y_pred_decision_o))
print("Subsample Data:")
print(classification_report(y_test, y_pred_decision_sub))
print("Oversample Data:")
print(classification_report(y_test, y_pred_decision_over))
```

Valores agregados por weighted average, en orden, precision, recall y f1-score:

*   Original: 0.28, 0.31, 0.24
*   Subsampling: 0.19, 0.31, 0.22
*   Oversampling: 0.30, 0.30, 0.26

###Comparación y conclusión

Los resultados obtenidos no son positivos, aunque son mejores que un clasificador aleatorio, que generaría un resultado de aproximadamente 1/7=0.14 (pues son 7 clases). El modelo de SVC con datos originales resutla ser el con mayor precision, recall y f1-score (simultáneamente). Se puede visualizar su matriz de confusión.


```python
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.metrics import confusion_matrix
from sklearn.utils.multiclass import unique_labels

cm = confusion_matrix(y_test, y_pred_SVC_o)

classes = unique_labels(y_test, y_pred_SVC_o)

df = pd.DataFrame(cm, index=classes, columns=classes)

g = sns.heatmap(df, annot=True, cmap="Blues")
g.set_yticklabels(g.get_yticklabels(), rotation=0)

plt.title('Confusion matrix \n')
plt.xlabel('Predicted label')
plt.ylabel('True label')

plt.autoscale()
```

La diagonal de matriz de confusión está solo ligeramente marcada y los valores de valores de precision, recall y F1 cercanos a solamente el 30-40%. Con respecto a la pregunta, hasta ahora, se puede concluir que los aspectos técnicos no son muy determinantes en su género.
